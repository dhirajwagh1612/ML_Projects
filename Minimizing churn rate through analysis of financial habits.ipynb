```python
df = dataset[['housing','is_referred','app_downloaded','web_user','app_web_user','ios_user','android_user','registered_phones',
                  'payment_type','waiting_4_loan','cancelled_loan','received_loan','rejected_loan','zodiac_sign','left_for_two_month_plus',
                  'left_for_one_month','is_referred']]
```

```python
fig = plt.figure(figsize = (70,35))
plt.suptitle('pie charat distribution', fontsize = 20)
for i in range(1, df.shape[1]+1):
    plt.subplot(6,3,i)
    f=plt.gca()
    f.axes.get_yaxis().set_visible(False)
    f.set_title(df.columns.values[i -1])
    values = df.iloc[:, i -1].value_counts(normalize = True).values
    index = df.iloc[:, i - 1].value_counts(normalize = True).index
    plt.pie(values, labels=index,autopct='%1.1f%%')
    plt.axis('equal')
    plt.tight_layout(rect = [0,0.03,1,0.])
```
