```python
df['origin'] = df['origin'].replace({1: 'america', 2: 'europe', 3: 'asia'})
df = pd.get_dummies(df, columns=['origin'])
df = df.replace('?', np.nan)
df = df.apply(lambda x: x.fillna(x.median()),axis=0)
```

```python
df
```

```python
df = df.replace('?', np.nan)
#df = df.drop('origin',axis = 1)
X = df.drop('mpg', axis=1)
y = df[['mpg']]
```

```python
df
```

```python
from sklearn import preprocessing
X_scaled = preprocessing.scale(X)
X_scaled = pd.DataFrame(X_scaled, columns=X.columns)
y_scaled = preprocessing.scale(y)
y_scaled = pd.DataFrame(y_scaled, columns=y.columns)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled,y_scaled, test_size=0.20, random_state=25)
```

```python
from sklearn.linear_model import LinearRegression
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)

for idx, col_name in enumerate(X_train.columns):
    print("The coefficient for {} is {}".format(col_name, linear_regressor.coef_[0][idx]))
```

```python
linear_regressor.score(X_scaled, y_scaled)
```

```python
intercept = linear_regressor.intercept_[0]

print("The intercept for our model is {}".format(intercept))
```

```python
from sklearn.linear_model import Ridge
ridge = Ridge(alpha=.3)
ridge.fit(X_train,y_train)
print ("Ridge model:", (ridge.coef_))
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso(alpha=0.1)
lasso.fit(X_train,y_train)
print ("Lasso model:", (lasso.coef_))
```

```python
print(linear_regressor.score(X_train, y_train))
print(linear_regressor.score(X_test, y_test))
```

```python
print(ridge.score(X_train, y_train))
print(ridge.score(X_test, y_test))
```

```python
print(lasso.score(X_train, y_train))
print(lasso.score(X_test, y_test))
```
